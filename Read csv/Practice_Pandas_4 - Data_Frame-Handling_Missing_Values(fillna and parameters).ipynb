# fillna()
    -Yo can pass int values in fillna()
    -int number 
    - dictionary: for string and int for number
    
    
 

# Parameter:
    -value
    - Method
    - axis
    - limit
    - inplace

# Method: 
    - it is the method to fill the values in the place of nan values
- bfill : values fill on the bases of the bottoms values into nan values
- backfill: equal to bfill
- ffill : values fill on the bases of the top values into nana values.
- pad: equal to ffill
-

```python
data.fillna(method="bfill")
```

```python
data.fillna(method="pad")
```

```python
data.fillna(method="backfill")
```

# Axis:
    - index 0 rows
    - columns 1
    - with axis should be pass the method value

```python
data.fillna(axis=0,method="bfill")
```

```python
data.fillna(axis=1,method="bfill")
```

# limit: 
    - it is use to how many values wants to be put in the place of nan values..
    - should pass int number  , method

```python
data.fillna(5,limit=5)  # with int
```

```python
data.fillna(method="bfill",limit=5)  # with method
```
