# Pandas Group By Function
- pandas group by function use to split the data into group based on some criteria

```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('D:\\PIAIC DOC\\Q2\\pandas\\PRACTICE\\Read CSV\\data.csv')
data
```

# Parameters:

# By:
    -Ye value ko break krta hai or object return krata ha.. or har value
    ki index ka pata chal jata hai
    

```python
by=data.groupby(by="Mid")  # object bn gye
```

```python
by.groups
```

```python
by1=data.groupby(by="Degree")
```

```python
by1.groups
```

```python
by2=data.groupby(by="Student_Name")
```

```python
by2.groups
```

```python
a=data.groupby(["Student_Name","Mid"])
```

```python
a.groups
```

```python
for i, j in a:
    print(i)
    print(j)
```

```python
#convert into list
list(a)
```

# get_group():
    - is me group by function  bus us values ko lata hai jo is me defined hoti hain.

```python
d1=data.groupby("Degree").get_group("BS(CS)")
d1
```

# axis : 
    -{0 or 'index', 1 or 'columns'}, 
    - default 0
    - Split along rows (0) or columns (1).

```python
mid=data.groupby(("Mid"),axis=1)  # columns wise split...
```

```python
mid.groups
```

```python
mid=data.groupby(("Mid"),axis=0)  # columns wise split...
```

```python
mid.groups
```
